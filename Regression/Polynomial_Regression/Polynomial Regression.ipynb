```python
# Polynomial Regression

# Importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Importing the dataset
dataset = pd.read_csv('/home/forbidden_devil/Machine Learning A-Z/Part 2 - Regression/Section 6 - Polynomial Regression/Position_Salaries.csv')
# We can use "X = dataset.iloc[:, 1].values" but this would make X an array
# It's always better to have X as a matrix and y as an array
X = dataset.iloc[:, 1:2].values
y = dataset.iloc[:, 2].values

# We don't make a train-test split here as the dataset is too small and we can get better results without splitting it
```

```python
# Fitting Linear Regression to the dataset
from sklearn.linear_model import LinearRegression
linreg1 = LinearRegression()
linreg1.fit(X, y)
```

```python
# Fitting Polynomial Regression to the dataset
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree = 2)
#poly_reg = PolynomialFeatures(degree = 3)
#poly_reg = PolynomialFeatures(degree = 4)
# X_poly has additional polyomial terms included
X_poly = poly_reg.fit_transform(X)
poly_reg.fit(X_poly, y)
linreg2 = LinearRegression()
linreg2.fit(X_poly, y)
```

```python
# Visualizing the Linear Regression results
plt.scatter(X, y, color = 'red')
plt.plot(X, linreg1.predict(X), color = 'blue')
plt.title("Linear Regression")
plt.xlabel("Position level")
plt.ylabel("Salary")
plt.show()
```

```python
# Visualizing the Polynomial Regression results of degree 2
plt.scatter(X, y, color = 'red')
plt.plot(X, linreg2.predict(poly_reg.fit_transform(X)), color = 'blue')
plt.title("Polynomial Regression")
plt.xlabel("Position level")
plt.ylabel("Salary")
plt.show()
```

```python
# Visualizing the Polynomial Regression results of degree 3
plt.scatter(X, y, color = 'red')
plt.plot(X, linreg2.predict(poly_reg.fit_transform(X)), color = 'blue')
plt.title("Polynomial Regression")
plt.xlabel("Position level")
plt.ylabel("Salary")
plt.show()
```

```python
# Visualizing the Polynomial Regression results of degree 4
plt.scatter(X, y, color = 'red')
plt.plot(X, linreg2.predict(poly_reg.fit_transform(X)), color = 'blue')
plt.title("Polynomial Regression")
plt.xlabel("Position level")
plt.ylabel("Salary")
plt.show()
```

```python
# Making the graph more smoother
# Visualizing the Polynomial Regression results of degree 4
X_grid = np.arange(min(X), max(X), 0.1)
X_grid = X_grid.reshape(len(X_grid), 1)
plt.scatter(X, y, color = 'red')
plt.plot(X_grid, linreg2.predict(poly_reg.fit_transform(X_grid)), color = 'blue')
plt.title("Polynomial Regression")
plt.xlabel("Position level")
plt.ylabel("Salary")
plt.show()
```

```python
# Predicting a new result with Linear Regression
linreg1.predict(6.5)
```

```python
# Predicting a new result with Polynomial Regression
linreg2.predict(poly_reg.fit_transform(6.5))
```
